```python
#Three conditions with conditional statement with if 
#loop that iterates 

for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0: #modulo operator % & each value of i in the loop is treated as an integer
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# Use dictionary. Itâ€™s flexible if you want to add more conditions 

fizzbuzz_dict = {3: "Fizz", 5: "Buzz", 7: "Bazz"}

for i in range(1, 101):
    output = ""
    for key, value in fizzbuzz_dict.items():
        if i % key == 0:
            output += value
    print(output or i)

# keys are divisors (3, 5 and 7) and values are the corresponding words ("Fizz", "Buzz", "Bazz").
#We check each divisor and add the corresponding word to output if i is divisible by that key.
#If output is empty (i.e., not divisible by 3 or 5), we print i; otherwise, we print output.
```
